# File I/O

### Opening and closing a file

```python
from Module_3_Notes import employees

myfile = open('test.txt', 'w')
myfile.write('Hello Python!')
myfile.close()
```

This created a new file called "test.txt" in the same directory.

Opening and automatically closing a text file (these next two examples are the preferred way)

```python
with open('test.txt', 'w') as myfile:
    myfile.write('hello Python!')
```

Opening a text file in read mode and automatically closing it

```python
with open('test.txt', 'r') as infile:
    print(infile.readline())
```

Writing one line to a text file

```python
with open('employees.txt', 'w') as file:
    file.write('Sally Smith\n')
```

Appending one line to a text file

```python
with open('employees.txt', 'a') as file:
    file.write('Mark Jacobson\n')
```

Using a for statement to read each line of the file. Also, notice that we don't need the mode 'r' added here because it defaults to read if not specified.

```python
with open('employees.txt') as file:
    for line in file:
        print(line, end='')
    print()
```

Reading the entire file as a string

```python
with open('employees.txt') as file:
    contents = file.read()
    print(contents)
```

Reading the entire file as a list. Notice that the readlines() stores the file as a list called employees.

```python
with open('employees.txt') as file:
    employees = file.readlines()
    print(employees[0], end='')
    print(employees[1])
```

Pretty cool, check out this way I figured out how to do the above method if there are tons of employees:

```python
with open('employees.txt') as file:
    employees = file.readlines()
    for employee in employees:
        print(employee, end='')
```

Reading each line of the file

```python
with open('employees.txt') as file:
    employee1 = file.readline()
    print(employee1, end='')
    employee2 = file.readline()
    print(employee2)
```

### Working with lists in a text file

Writing items in a list to a text file

In this example, two names are stored within an employees list. Then, a for loop uses the write() method of the file object to write each member in the list to the file. To make that work, this code adds a enw line character (\n) to the end of the string for each member.

```python
employees = ['Sally Smith', 'Mark Jacobson']
with open('employees.txt', 'w') as file:
    for emp in employees:
        file.write(emp + '\n')
```

Reading the lines in a file into a list

```python
employees = []
with open('employees.txt') as file:
    for line in file:
        line = line.replace('\n', '')
        employees.append(line)
print(employees)
```

Reading and writing a list of numbers (converting to a str)

```python
extensions = [4233, 5443, 5221]
with open('extensions.txt', 'w') as ext_file:
    for ext in extensions:
        ext_file.write(str(ext) + '\n')
```

And converting the string back to integers

```python
extensions = []
with open('extensions.txt') as file:
    for line in file:
        line = line.replace('\n', '')
        extensions.append(int(line))
print(extensions)
```

### Writing a list to a CSV file

To write data to a CSV file, you use the writer() function of the csv module to get a writer object. Then, you use the methods of the writer object to write data.

```python
import csv
employees = [['Sally Smith', 5635],['Mark Jacobson', 6436],['John Staton', 7436]]
with open('employees.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(employees)
```

From line 2 above: if you exclude the newline parameter there it may not work on all operating systems

Reading from a CSV file

```python
import csv
with open('employees.csv', newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row[0] + ' (x' + str(row[1]) + ')')
```

'x' is for extension
