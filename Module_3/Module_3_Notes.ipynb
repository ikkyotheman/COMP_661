Since Python 3.7x the order of key-value pairs is the same as their input order, i.e. dictionaries are now ordered structures.

But that doesn’t really matter because the elements of a dictionary are never indexed with integer indices. Instead, you use the keys to look up the corresponding values:

```python
eng2sp['two']
```

```python
eng2sp['four'] # A Key error is what arises if the key isn't in the dictionary
```

The length function can be used on dictionaries

```python
len(eng2sp)
```

The in operator works on dictionaries; it tells you whether something appears as a key in the dictionary (appearing as a value is not good enough).

```python
'uno' in eng2sp
```

To see whether something appears as a value in a dictionary, you can use the method values, which returns the values as a type that can be converted to a list, and then use the in operator:

```python
vals = list(eng2sp.values())
print(type(vals))
'uno' in vals
```

```python
string = "Brontosaurus"
d = dict()
for c in string:
    if c not in d:
        d[c] = 1
    else:
        d[c] += 1
d
```

Dictionaries have a method called get that takes a key and a default value. If the key appears in the dictionary, get returns the corresponding value; otherwise it returns the default value ('0' in this example but it could be a string like 'not found'). For example:

```python
counts = { 'chuck' : 1 , 'annie' : 42, 'jan': 100}
print(counts.get('jan',0))
```

This is the same histogram loop rewritten more concisely with the get() method.

```python
string = "Brontosaurus"
d = dict()
for c in string:
    d[c] = d.get(c, 0) + 1
print(d)
```

Example from PY4E - files using get()

```python
fname = input('What\'s the file name?')
fhand = open(fname)
fhand = fhand.read()
fhand = fhand.rstrip()
combo = dict()

for letter in fhand:
    combo[letter] = combo.get(letter, 0) + 1
print(combo)
```

# Dictionary of Dictionaries

We have a bunch of servers and we are trying to see if some are online or offline

```python
infrastructure = {
    "server_01" : {"ip" : "192.168.1.1" , "status" : "online"},
    "server_02" : {"ip" : "192.168.1.2" , "status" : "offline"}
}
```

```python
infrastructure = {
    "server_01" : {"ip" : "192.168.1.1" , "status" : "online"},
    "server_02" : {"ip" : "192.168.1.2" , "status" : "offline"}
}

print(infrastructure)
print(infrastructure["server_02"]["status"])
infrastructure.get("server_02").get("status")
infrastructure.get("server_03", "not found")
infrastructure.get("server_03", {}).get("status", "unknown server")
```
