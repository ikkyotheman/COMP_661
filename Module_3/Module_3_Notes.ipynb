This is 2 key valued pairs below

```python
user_data = {
  "id": 101,
  "role": "Admin",
  "status": "Active"
}
keys_view = user_data.keys()
print(keys_view)
vals_view = user_data.values()
print(vals_view)
for key, value in user_data.items():
  print(f"Attribute: {{key}} -> Value : {{value}}")
print()
other_user_keys = {"id", "email", "ip_address"}
shared_fields = user_data.keys() & other_user_keys
print(f"Common fields: {shared_fields}")
```

```python
keys_view = keys_view.keys()
print(keys_view)
vals_view = user_data.values()
print(vals_view)
```

List of Dictionaries

```python
employees = [
    {"name" : "Steven" , "dept" : "Engineering"},
    {"name" : "Elise" , "dept" : "Design"}
]

# Get the Department of Elise
print(employees[1].get('department', 'Not Found'))
print(employees[1].get('dept', 'Not Found'))
```

# Dictionary of Dictionaries

We have a bunch of servers and we are trying to see if some are online or offline

```python
infrastructure = {
    "server_01" : {"ip" : "192.168.1.1" , "status" : "online"},
    "server_02" : {"ip" : "192.168.1.2" , "status" : "offline"}
}
```

```python
infrastructure = {
    "server_01" : {"ip" : "192.168.1.1" , "status" : "online"},
    "server_02" : {"ip" : "192.168.1.2" , "status" : "offline"}
}

print(infrastructure)
print(infrastructure["server_02"]["status"])
infrastructure.get("server_02").get("status")
infrastructure.get("server_03", "not found")
infrastructure.get("server_03", {}).get("status", "unknown server")
```
