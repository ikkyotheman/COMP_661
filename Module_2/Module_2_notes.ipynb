This is an example of how an int (goes for other variables like strings and tuples) are immutable, value1 will always be 25 and value2 will always be 50.

```python
def double_the_number(value):
    value = value * 2
    return value
value1 = 25
value2 = double_the_number(value1)
print(value1)
print(value2)
```

count(item) - Returns the number of occurrences of an item in the list. If the item isn't found in the list, this method return 0.
reverse(list) - Reverses the order of the items in the list.
sort([key=function]) - Sorts the list items in place. The optional key argument specifies a function to be called on each item before sorting.
sorted(list[, key=function]) - Returns a employee list consisting of the sorted items of the orginal list. The optional key argument specifies a function to be called on each time before sorting.

```python
numlist = [5, 15, 84, 3, 14, 2, 8, 10, 14, 25]
print(numlist.count(14))
print(len(numlist))
print(numlist.reverse())
print(numlist.sort())
```

```python
foodlist = ['orange', 'apple', 'Pear', 'banana']
sorted_foodlist = sorted(foodlist, key=str.lower)
print(sorted_foodlist)
print(foodlist)
```

How to make a shallow copy of a list (this is because lists are mutable types)

```python
list_one = [1, 2, 3, 4, 5]
list_two = list_one
list_one[1] = 4
print(list_one)
print(list_two)
```

```python
import copy
list_one = [1, 2, 3, 4, 5]
list_two = copy.deepcopy(list_one)
list_two[1] = 4
print(list_one)
print(list_two)
```

```python
def display_menu():
    print('Welcome to the Employee Management Program')
    print()
    print('COMMAND MENU')
    print('show - Show all employees')
    print('add - Add an employee')
    print('del - delete an employee')
    print('exit - Exit the program')
    print()

def display(employee_list):
    if len(employee_list) == 0:
        print('There are no employees in the list.\n')
        return
    else:
        i = 1
        for employee in employee_list:
            print(str(i) + '. ' + employee[0] + '  ('+ employee[1] + ')')
            i += 1
        print()

def add(employee_list):
    name = input('Enter the employee\'s name:')
    title = input('Enter the employee\'s title:')
    years = input('Years with company: ')
    employee = []
    employee.append(name)
    employee.append(title)
    employee.append(years)
    employee_list.append(employee)
    print(employee[0] + 'was added to the list \n')

def delete(employee_list):
    number = int(input('Enter which number to delete:'))
    if number < 1 or number > len(employee_list):
        print('Please enter a valid number')
    else:
        employee = employee_list.pop(number-1)
        print(employee[0] + ' was removed from the list \n')
def main():
    employee_list = [['Peter Kindrachuk', 'President', 10],
                     ['Heather Kindrachuk', 'CEO', 9],
                     ['Master Jin', 'Grandmaster', 30],
                     ['Roy Orbison', 'Producer', 30]]
    display_menu()
    while True:
        command = input('Enter a command: ')
        if command.lower() == 'show':
            display(employee_list)
        elif command.lower() == 'add':
            add(employee_list)
        elif command.lower() == 'del':
            delete(employee_list)
        elif command.lower() == 'exit':
            print('Exiting the program...')
            break
        else:
            print('That is not a valid command. Please try again. \n')

if __name__ == '__main__':
    main()
```

```python
import random

def crunch_numbers(data):
    total = 0
    for number in data:
        total += number

    average = round(total / len(data))
    median_index = len(data) // 2
    median = data[median_index]
    minimum =  min(data)
    maximum = max(data)
    dups = get_duplicates(data)

    print('Average=', average,
          'Median=', median,
          'Minimum=', minimum,
          'Maximum=', maximum,
          'Duplicates = ', dups)

def get_duplicates(data):
    dups = []
    for i in range (51):
        count = data.count(i)
        if count >= 2:
            dups.append(i)
    return dups

def main():
    fixed_tuple = (0,5,10,15,20,25,30,35,40,45,50)
    random_list = [0] * 11
    for i in range(len(random_list)):
        random_list[i] = random.randint(0, 50)
    random_list.sort()

    print('TUPLE DATA:', fixed_tuple)
    crunch_numbers(fixed_tuple)
    print()
    print('RANDOM DATA:', random_list)
    crunch_numbers(random_list)


if __name__ == '__main__':
    main()
```
