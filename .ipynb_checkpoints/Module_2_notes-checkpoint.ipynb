### Module 2, Video 5

Passing lists into functions

This is an example of how an int (goes for other variables like strings and tuples) are immutable, value1 will always be 25 and value2 will always be 50.

```python
def double_the_number(value):
    value = value * 2
    return value
value1 = 25
value2 = double_the_number(value1)
print(value1)
print(value2)
```

Here is an example with a list which is immutable.

```python
def add_to_list(mylist, item):
    mylist.append(item)
inventory = ['jeans', 'hat', 'shirt']
add_to_list(inventory, 'socks')
print(inventory)
```

How to define a list of lists with 3 rows and 3 columns

```python
movies = [['The Holy Grail', 1975, 9.99], 
          ['Life of Brian', 1979, 12.30],
          ['The Meaning of Life', 1983, 7.50]]
```

This is the example that shows how you can programmatically add itemns to a nested list:

```python
movies = [['The Holy Grail', 1975],
         ['Life of Brian', 1979]]
movie = []                           # Create empty list for next movie
movie.append('The Meaning of Life')  # Add movie name to movie list
movie.append(1983)                   # Add movie year to movie list
movies.append(movie)                 # Add movie list to movies list
print(movies)
```

```python
movies = [['The Holy Grail', 1975, 9.99], 
          ['Life of Brian', 1979, 12.30],
          ['The Meaning of Life', 1983, 7.50]]
print(movies[0][0])
print(movies[0][2])
print(movies[1][2])
print(movies[1][0])
print(movies[2][0])
```

```python
movies = [['The Holy Grail', 1975, 9.99], 
          ['Life of Brian', 1979, 12.30],
          ['The Meaning of Life', 1983, 7.50]]
for movie in movies:
    for item in movie:
        print(item, end=' | ')
```

count(item) - Returns the number of occurrences of an item in the list. If the item isn't found in the list, this method return 0.
reverse(list) - Reverses the order of the items in the list.
sort([key=function]) - Sorts the list items in place. The optional key argument specifies a function to be called on each item before sorting.
sorted(list[, key=function]) - Returns a new list consisting of the sorted items of the orginal list. The optional key argument specifies a function to be called on each time before sorting.

```python
numlist = [5, 15, 84, 3, 14, 2, 8, 10, 14, 25]
print(numlist.count(14))
print(len(numlist))
print(numlist.reverse())
print(numlist.sort())
```

How to sort a list using the built in sort function: 

```python
foodlist = ['orange', 'apple', 'Pear', 'banana']
print(foodlist)
foodlist.sort(key=str.lower)
print(foodlist)
```

How to use the built in sorted function to return a sorted list to a different variable:

```python
foodlist = ['orange', 'apple', 'Pear', 'banana']
sorted_foodlist = sorted(foodlist, key=str.lower)
print(sorted_foodlist)
print(foodlist)
```

How to make a shallow copy of a list (this is because lists are mutable types)

```python
list_one = [1, 2, 3, 4, 5]
list_two = list_one
list_one[1] = 4
print(list_one)
print(list_two)
```

```python
import copy
list_one = [1, 2, 3, 4, 5]
list_two = copy.deepcopy(list_one)
list_two[1] = 4
print(list_one)
print(list_two)
```

```python

# Slicing a list
# syntax:

mylist[start:end:step]
```

```python
numbers = [52, 54, 56, 58, 60, 62]
print(numbers[0:2])
print(numbers[:2])
print(numbers[4:])
```

```python
numbers = [52,54, 56, 58, 60, 62]
print(numbers[0:4:2])
print(numbers[::-1])
```

### Employee Management Program

This is starting the assigment

```python
import random

alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's','t' ,'u', 'v', 'w', 'y', 'x', 'z']
i = 0
alpha = []
while i < 10:
    letter = random.choice(alphabet)
    alpha.append(letter)
    i += 1
print(alpha)
alpha_sorted = sorted(alpha) 
print(alpha_sorted)
alpha_rev_slice = alpha_sorted[::-1] # If you feel like a slice reverse
print(alpha_rev_slice)
alpha_sorted.reverse()               # Using the 'reverse' method (returns 'None')
print(alpha_sorted)
```
